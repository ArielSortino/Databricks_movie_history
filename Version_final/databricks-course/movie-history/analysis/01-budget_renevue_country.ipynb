#### Vamos a calcular trabajar sobre el presupuesto y los ingresos, tenemos que calcular el total y el promedio de ambos, esto por país y de las películas lanzadas entre el 2010 y el 2015, todo ordenado del país de mayores ingresos al de menos ingresos, además ponemos el total de peliculas sobre lo que se calcula esto


```sql
SELECT country_name,
       COUNT(country_name) AS total_movie,
       SUM(budget) AS total_budget,
       CAST(AVG(budget) AS DECIMAL(18,2)) AS avg_budget,
       SUM(revenue) AS total_revenue,
       CAST(AVG(revenue) AS DECIMAL(18,2)) AS avg_revenue
FROM movie_gold.results_movie
WHERE year_release_date BETWEEN 2010 AND 2015
GROUP BY country_name
ORDER BY total_revenue DESC
```
