### Configurar Acceso Azure Data Lake Storage Gen2
1. Crear el objeto "External Location" llamado "external_location_bronze_mov_his"
2. Crear el objeto "External Location" llamado "external_location_silver_mov_his"
3. Crear el objeto "External Location" llamado "external_location_gold_mov_his"
4. Validar el acceso a los objetos "External Location" creados

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_bronze_mov_his
    URL "abfss://bronze@moviee.dfs.core.windows.net/"
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For The Bronze Data Lakehouse'








```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_silver_mov_his
    URL "abfss://silver@moviee.dfs.core.windows.net/"
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For The Silver Data Lakehouse'
```

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_gold_mov_his
    URL "abfss://gold@moviee.dfs.core.windows.net/"
    WITH (STORAGE CREDENTIAL storage_credential_mov_his)
    COMMENT 'External Location For The Gold Data Lakehouse'
```
