### Acceder a Azure Data Lake Storage mediante Token SAS
1. Establecer la configuración de spark "SAS Token"
2. Listar archivos del contenedor "demo"
3. Leer datos del archivo "movie.csv"

```python
spark.conf.set("fs.azure.account.auth.type.moviee.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.moviee.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.moviee.dfs.core.windows.net", "sp=rl&st=2026-03-12T12:52:05Z&se=2026-03-12T21:07:05Z&spr=https&sv=2024-11-04&sr=c&sig=mv1vTN3UhnAvupzt7kUQJofmFI3J1JH84RKW%2BtxAgj0%3D")






























```

```python
display(dbutils.fs.ls("abfss://demo@moviee.dfs.core.windows.net/"))
```
