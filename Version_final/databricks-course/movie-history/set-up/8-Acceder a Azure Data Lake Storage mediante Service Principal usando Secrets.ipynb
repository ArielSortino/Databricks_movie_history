### Acceder a Azure Data Lake Storage mediante Service Principal usando Secrets
##### Este notebook es un clon del notebook 3-Azure Data Lake Storage mediante Service Principal, la única diferencia es que vamos a usar la utility Secrets para obtener la key que pasamos en la configuración del spark en vez de hardcodearla

```python
client_id = dbutils.secrets.get(scope= 'movie-history-secret-scope', key= 'client-id')
tenant_id = dbutils.secrets.get(scope= 'movie-history-secret-scope', key= 'tenant-id')
client_secret = dbutils.secrets.get(scope= 'movie-history-secret-scope', key= 'client-secret')







```

```python
spark.conf.set("fs.azure.account.auth.type.moviee.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.moviee.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.moviee.dfs.core.windows.net", client_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.moviee.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.moviee.dfs.core.windows.net", f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
























```

Por ultimo, tenemos que asignar el rol "Storage Blob Data Contributor" al Data Lake para leer la info, para eso vamos a Storage-account en el portal de azure, voy a moviee, voy a Access Control (IAM), vamos a Add, Add Role Assignment y ponemos Storage Blob Data Contributor, despues next y en members ponemos movie-app, despues todo next y creamos el rol
