### Acceder a Azure Data Lake Storage mediante Access Key usando Secrets
#####Este notebook es un clon del notebook 1-Acceder a Azure Data Lake Storage mediante Access Key, la única diferencia es que vamos a usar la utility Secrets para obtener la key que pasamos en la configuración del spark en vez de hardcodearla

```python
movie_access_key = dbutils.secrets.get(scope= 'movie-history-secret-scope', key= 'movie-access-key')






```

```python
spark.conf.set(
    "fs.azure.account.key.moviee.dfs.core.windows.net",
    movie_access_key
)










```

```python
dbutils.fs.ls("abfss://demo@moviee.dfs.core.windows.net/")












```

```python
display(dbutils.fs.ls("abfss://demo@moviee.dfs.core.windows.net/"))
```

```python
spark.read.csv("abfss://demo@moviee.dfs.core.windows.net/movie.csv")
```

```python
display(spark.read.csv("abfss://demo@moviee.dfs.core.windows.net/movie.csv"))
```
