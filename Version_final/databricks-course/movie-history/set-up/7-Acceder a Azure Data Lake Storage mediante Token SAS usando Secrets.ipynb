### Acceder a Azure Data Lake Storage mediante Token SAS usando Secrets
##### Este notebook es un clon del notebook 1-Acceder a Azure Data Lake Storage mediante Token SAS, la única diferencia es que vamos a usar la utility Secrets para obtener la key que pasamos en la configuración del spark en vez de hardcodearla

```python
movie_sas_token = dbutils.secrets.get(scope= 'movie-history-secret-scope', key= 'movie-sas-token')






```

```python
spark.conf.set("fs.azure.account.auth.type.moviee.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.moviee.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.moviee.dfs.core.windows.net", movie_sas_token)





























```

```python
display(dbutils.fs.ls("abfss://demo@moviee.dfs.core.windows.net/"))
```

```python
display(spark.read.csv("abfss://demo@moviee.dfs.core.windows.net/movie.csv"))
```
