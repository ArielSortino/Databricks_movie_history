### Aggregate Functions (Funciones de agregación)

#### Se puede ver más info en https://spark.apache.org/docs/latest/api/sql/index.html#max

```sql
SELECT COUNT(1)
FROM movies
```

```sql
SELECT MAX(release_date)
FROM movies
```

```sql
SELECT year_release_date, count(1) AS total_count, SUM(budget) AS sum_budget, 
MAX(budget) AS max_budget, MIN(budget) AS min_budget, AVG(budget) AS avg_budget
FROM movies
WHERE year_release_date is not null
GROUP BY year_release_date
HAVING count(1) > 220
ORDER BY year_release_date DESC
```

### Windows Functions (Funciones de ventana)

#### Se puede ver más info en https://spark.apache.org/docs/latest/api/sql/index.html#rank

```sql
SELECT title, year_release_date, release_date,
      RANK() OVER(PARTITION BY year_release_date ORDER BY release_date) as rank
FROM MOVIES
WHERE year_release_date is not null
ORDER BY release_date
```
