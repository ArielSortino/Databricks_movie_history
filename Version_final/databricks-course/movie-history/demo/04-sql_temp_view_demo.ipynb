### Acceso al DataFrame mediante SQL

### Local Temp View
1. Crear vista(view) temporal en la base de datos
2. Acceder a la vista(view) desde la celda "SQL"
3. Acceder a la vista(view) desde la celda "Python"

#### Crear vista(view) temporal en la base de datos

```python
%run "../includes/configuration"
```

```python
results_movie_genre_language_df = spark.read.parquet(f"{gold_folder_path}/results_movie_genre_language")
```

#### Acceder a la vista(view) desde la celda "SQL"

```python
results_movie_genre_language_df.createTempView("v_movie_genre_language")




```

```python
%sql
SELECT * FROM v_movie_genre_language WHERE VOTE_AVERAGE > 7.5
```

#### Acceder a la vista(view) desde la celda "Python"

```python
results_movie_genre_language_2 = spark.sql("SELECT * FROM v_movie_genre_language WHERE VOTE_AVERAGE > 7.5")




```

```python
display(results_movie_genre_language_2)
```

```python
p_average = 7.5




```

```python
display(spark.sql(f"SELECT * FROM v_movie_genre_language WHERE VOTE_AVERAGE > {p_average}"))
```

### Global Temp View
1. Crear vista(view) temporal en la base de datos
2. Acceder a la vista(view) desde la celda "SQL"
3. Acceder a la vista(view) desde la celda "Python"


```python
results_movie_genre_language_df.createOrReplaceGlobalTempView("gv_movie_genre_language")



```

#### Acceder a la vista(view) desde la celda "SQL"

```python
%sql
SELECT * FROM gv_movie_genre_language WHERE VOTE_AVERAGE > 7.5
```

```python
%sql
SHOW TABLES;
```

```python
%sql
SHOW TABLES IN global_temp;
```

```python
%sql
SELECT * FROM global_temp.gv_movie_genre_language WHERE VOTE_AVERAGE > 7.5
```

#### Acceder a la vista(view) desde la celda "Python"

```python
spark.sql("SELECT * FROM global_temp.gv_movie_genre_language WHERE VOTE_AVERAGE > 7.5").display()
```
